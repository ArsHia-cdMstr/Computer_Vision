```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
import main as nois 
```

```python
def plot_image_vs_noisyImg(original_image: np.ndarray, noisy_input_image: list[np.ndarray], noise_type:list[str]):
    # تصویر نهایی را نمایش می‌دهیم با استفاده از matplotlib
    plt.figure(figsize=(15,15))
    plt.subplot(321), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
    plt.subplot(323), plt.imshow(noisy_input_image[0], cmap='gray'), plt.title(f'Image with {noise_type[0]} Noise with {noise_type[1]}')
    plt.subplot(324), plt.imshow(noisy_input_image[1], cmap='gray'), plt.title(f'Image with {noise_type[0]} Noise with {noise_type[2]}')
    plt.subplot(325), plt.imshow(noisy_input_image[2], cmap='gray'), plt.title(f'Image with {noise_type[0]} Noise with {noise_type[3]}')
    plt.subplot(326), plt.imshow(noisy_input_image[3], cmap='gray'), plt.title(f'Image with {noise_type[0]} Noise with {noise_type[4]}')
    plt.show()
```

```python
# تصویر اولیه را بارگیری می‌کنیم
image = cv2.imread('DS/Med/1.png', cv2.IMREAD_GRAYSCALE)
```

# Gaussian Noise

در تصویر زیر شما نویز گوسین با پارامتر های داده شده میسازیم و میبینیم که 

```python
# نویز Gaussian را به تصویر اضافه می‌کنیم
parameter_for_gaussian = [(0,40), (4, 30), (10, 0), (10 , 25)]
noise_type_for_gaussian = ["Gaussian", "mean=0, std=1", "mean=4, std=30", "mean=10, std=0", "mean=10, std=25"]
image_noisy_set = []

for parameter in parameter_for_gaussian:
    noisy_image = nois.add_gaussian_noise(image, mean=parameter[0], std=parameter[1])
    image_noisy_set.append(noisy_image)
    
plot_image_vs_noisyImg(image, image_noisy_set, noise_type=noise_type_for_gaussian)
```

# salt and pepper noise

در تصویر زیر ما عکسی با تراکم 

```python
# نویز Gaussian را به تصویر اضافه می‌کنیم
noisy_image = nois.add_salt_pepper_noise(image, 0.2)
plot_image_vs_noisyImg(image, noisy_image, noise_type='Salt and Pepper')
```
