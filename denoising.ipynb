```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def salt_pepper_noise(image, density):
    salt_prob = density
    pepper_prob = density
    output = np.zeros(image.shape, np.uint8)
    thres = 1 - salt_prob
    thres2 = 1 - salt_prob - pepper_prob
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = np.random.random()
            if rdn < pepper_prob:
                output[i][j] = 0
            elif rdn < thres2:
                output[i][j] = image[i][j]
            else:
                output[i][j] = 255
    return output


#  
image = cv2.imread('DS/Med/1.png', 0)  #         

#     
noisy_image = salt_pepper_noise(image, 0.05)  #     

#  
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[1].imshow(noisy_image, cmap='gray')
axs[1].set_title('Image with Salt and Pepper Noise')
plt.show()
```
